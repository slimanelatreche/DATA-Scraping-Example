# Extraction Function

```python
def Extract_Details(link):
    from selenium import webdriver
    from selenium.webdriver.common.keys import Keys
    import time
 
    driver = webdriver.Chrome()
    driver.get(link)

    time.sleep(4) #pause the code for 4 sec
    
    recompense = driver.find_element_by_xpath('//h1[@class="ng-binding"]').text
    points = driver.find_element_by_xpath('//div[@class="price ng-binding"]').text
    recompense_description = driver.find_element_by_xpath('//article[@class="ng-binding"]').text
    partenaire = driver.find_element_by_xpath('//div[@class="partner"]').text
    partenaire_description = driver.find_element_by_xpath('//div/article/div[@class="ng-binding"]').text
    siteweb = driver.find_element_by_xpath('//a[@class="ng-binding ng-scope"]').text
    
    driver.close()
    
    return recompense, points, recompense_description, partenaire, partenaire_description, siteweb
```
